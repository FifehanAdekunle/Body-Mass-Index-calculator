```python
# import lib

from bs4 import BeautifulSoup
import requests

import time
import datetime
import smtplib
```

```python
# connect to website for scraping
URL ='https://www.fashionnova.com/products/maeve-scoop-neck-snatched-jumpsuit-black'
# headers 
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL,headers=headers)

Soup1 =  BeautifulSoup(page.content,"html.parser")

print (Soup1)
```

```python
# prettify  adds structures beauty to your data
Soup2 =  BeautifulSoup(Soup1.prettify(), "html.parser")
print (Soup2)
```

```python
# pulling data 
title = Soup2.find(class_='product-info__title').get_text()
print(title)
```

```python
price = Soup2.find(class_="product-info__price product-info__price--sale").get_text()
print(price)
```
