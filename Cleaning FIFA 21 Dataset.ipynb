# 5. Extracting Contract Information

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index,row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print (row['Contract'])
```

```python
# Extracting contract information and creating new columns
def extract_contract_info(Contract):
    if Contract == 'Free'or'On Loan' in Contract:
            start_date = np.nan
            end_date = np.nan
            contract_length = 0
    else:
        start_date,end_date = Contract.split(' ~ ')
        start_year = int( start_date[:4])
        end_year = int( end_date[:4])
        contract_length = end_year - start_year
    return start_date,end_date,contract_length 

# Apply function to contract col and create new col
new_cols=['Contract Start','Contract End', 'Contract (years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i ,column =new_cols[i], value = new_data[i])
```

```python
fifa.sample(3)
```

# 6. Creating Contract Categories


```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract (years)']].head(5)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract (years)']].tail(20)
```

```python
# creating contract categorizes
def cat_contract(Contract):
    if Contract == 'Free':
        return  'Free'
    elif 'On loan' in Contract:
        return 'On loan'
    else:
        return 'Contract'
        
# insert new col
fifa.insert(fifa.columns.get_loc('Contract (years)')+1, 'Contract Status', fifa['Contract'].apply(cat_contract))
fifa.sample(5)           
```

# 7. Cleaning Player Heights

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
# Cleaning and converting player heights to centimeters
def convert_height(Height):
    if "cm" in Height:
        return int(Height.strip("cm"))
    else:
        feet,inches = Height.split("'")
        total_inches = int(feet)*12+ int(inches.strip('"'))
        return round(total_inches *2.54)
# Applying the function to the "Height" column and renaming the column
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'] .unique()
```

```python
# rename height col
fifa = fifa.rename(columns = {'Height': "Height(cm)"})
fifa.sample(1)
```

# 8. Cleaning Player Weights

```python
fifa ['Weight'].dtype
```

```python
fifa ['Weight'].unique()
```

```python
# Cleaning and converting player weights to kilograms
def convert_Weight(Weight):
    if "kg" in Weight:
        return int(Weight.strip("kg"))
    else:
        pounds = int(Weight.strip("lbs"))
        return round(pounds/2.205)
 # Applying the function to the "Weight" column and renaming the column
fifa['Weight'] = fifa['Weight'].apply(convert_Weight)
fifa['Weight'] .unique()
```

```python
# rename weight col
fifa = fifa.rename(columns = {'Weight': "Weight(kg)"})
fifa.sample(1)
```
